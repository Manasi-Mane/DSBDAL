```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab')
```

```python
text = """Text analytics is the process of analyzing text data to extract useful information."""
```

```python
tokens = word_tokenize(text)
```

```python
print(tokens)
```

```python
pos_tags = pos_tag(tokens)
print(pos_tags)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print('After stopword removal: ', filtered_tokens)
```

```python
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_tokens]
print('Stemmed Words: ', stemmed_words)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print('Lemmatized Words: ', lemmatized_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents = [
    "Text analytics is useful",
    "Text data analysis is important",
    "Analytics extracts information from text"
]
```

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
feature_names = vectorizer.get_feature_names_out()
```

```python
tfidf_array = tfidf_matrix.toarray()
```

```python
print("Feature Names:", feature_names)
print("TF-IDF Matrix:\n", tfidf_array)
```

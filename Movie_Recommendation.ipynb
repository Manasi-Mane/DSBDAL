```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load dataset
df = pd.read_csv("movie_dataset.csv")

# Display first 5 rows
print(df.head())
```

```python
print(df.columns)
```

```python
# Select features
features = ['keywords', 'cast', 'genres', 'director']
```

```python
# Fill missing values
for feature in features:
    df[feature] = df[feature].fillna('')
```

```python
# Combine features
def combine_features(row):
    return row['keywords'] + " " + row['cast'] + " " + row['genres'] + " " + row['director']

df["combined_features"] = df.apply(combine_features, axis=1)

print(df["combined_features"].head())
```

```python
# Convert text data to matrix
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
# Calculate cosine similarity
cosine_sim = cosine_similarity(count_matrix)
```

```python
# Helper functions
#Get movie index from title
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]
```

```python
#Get index from movie title
def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]
```

```python
# Take movie name from user
movie_user_likes = input("Enter your favorite movie: ")

# Get index of the movie from dataset
movie_index = get_index_from_title(movie_user_likes)

# Find similarity scores of all movies with selected movie
similar_movies = list(enumerate(cosine_sim[movie_index]))

# Sort movies based on similarity score in descending order
sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)

#Display Recommendations
# Display heading
print("\nTop Recommended Movies:\n")

# Counter variable
i = 0

# Loop through similar movies
for movie in sorted_similar_movies:

    # Print movie title using movie index
    print(get_title_from_index(movie[0]))

    # Increase counter
    i += 1

    # Stop after showing 10 recommendations
    if i > 10:
        break
```
